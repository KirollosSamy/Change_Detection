```python
import torch
from src.datasets.dataloader import create_data_loaders, create_test_loader
from src.models.FC_EF import FC_EF
from src.models.FC_Siamese import FC_Siamese
from src.training.trainer import train, validate
from src.training.evaluation import evaluate, test
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
model = FC_Siamese(in_channels=3, out_channels=1)
model = model.to(device)
```

```python
# Hyperparameters

batch_size = 16
val_ratio = 0.2

epochs = 60
lr = 0.001
gamma = 0.95
weight_decay = 1e-4
```

```python
data_dir = "data"
model_dir = "/content/drive/MyDrive/Change_Detection/models/FC_Siamese/batch_16_lr_0.001_gamma_0.95_decay_4_epochs_60"
```

```python
train_loader, val_loader = create_data_loaders(data_dir, val_ratio, batch_size, grayscale=False)
```

```python
train(model, train_loader, val_loader, epochs, model_dir, lr, gamma, weight_decay, device)
```

```python
# Load saved model

model_path = "/content/drive/MyDrive/Change_Detection/models/FC_Siamese/batch_16_lr_0.001_gamma_0.95_decay_4_epochs_60/model_59.pth"
state_dict = torch.load(model_path)
model.load_state_dict(state_dict)
```

```python
evaluate(model, val_loader, device, verbose=False)
```
